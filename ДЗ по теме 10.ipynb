```python
import pandas as pnd
import json
import matplotlib.pyplot as plt
import seaborn as sbn

# Загрузка данных
with open('events.json', 'r') as f:
   data = json.load(f)
#data = {"events": [ {"timestamp": "2023-08-21T08:00:00", "signature": "MALWARE-CNC Win.Trojan.Jadtre variant outbound connection"},] }

# Преобразование в DataFrame
df = pnd.DataFrame(data["events"])

# Визуализация
plt.figure(figsize=(10, 6))
sbn.countplot(data=df, x="signature")

plt.title("Распределение типов событий безопасности")
plt.xticks(rotation=90)  # Поворот подписей оси X для лучшей читаемости
plt.show()
```
